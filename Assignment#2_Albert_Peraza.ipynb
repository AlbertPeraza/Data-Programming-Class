```python
sql = ''' 
CREATE TABLE IF NOT EXISTS newsapi.sources (
    category varchar(30),
    country varchar(5),
    description varchar(500),
    id varchar(50),
    language varchar(5),
    name varchar(50),
    url varchar(100),
    PRIMARY KEY(name)
)
'''
engine.execute(sql)
```
